### Let's plot the shots map for the two teams

```python
pitchLengthX=120
pitchWidthY=8


home_team_required ="Barcelona"
away_team_required ="Manchester United"


pitch = Pitch()
fig, ax = pitch.draw(figsize=(120, 80))
    
for i,shott in shots.iterrows():
    pitchLengthX=120
    pitchWidthY=80
    goal=shott['shot_outcome']=='Goal'
    team_name=shott['team']
    
    circleSize=2
    
    x=shott['location'][0]
    y=shott['location'][1]
    
    
    
    if (team_name==home_team_required):    
    
        if goal:
            shotCircle=plt.Circle((x,pitchWidthY-y),circleSize,color="red")
            plt.text((x+1),pitchWidthY-y+1,shott['player'])
            
        else:
            shotCircle=plt.Circle((x,pitchWidthY-y),circleSize,color="red")     
            shotCircle.set_alpha(.2)
            
        ax.add_patch(shotCircle)
    elif (team_name==away_team_required):         
    
        if goal:
            shotCircle=plt.Circle((pitchLengthX-x,y),circleSize,color="blue") 
            plt.text((pitchLengthX-x+1),y+1,shott['player'])
        else:
            shotCircle=plt.Circle((pitchLengthX-x,y),circleSize,color="blue")      
            shotCircle.set_alpha(.2)
        ax.add_patch(shotCircle)
          
   
plt.text(5,75,away_team_required + ' shots') 
plt.text(80,75,home_team_required + ' shots') 
     
fig.set_size_inches(20, 14)
 
plt.show()
```

# Shot heatmap

```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch
```

```python
# Let's plot the shots made by Barcelona:
match_id_required = [18236]
team_required = "Barcelona"
```

```python
for ic,match_id in enumerate(match_id_required):
    
    #Load in all match events 
    
    team_actions = (bar_vs_manu['team']==team_required)
    
    #A dataframe of shots
    shots_match = bar_vs_manu[bar_vs_manu['type'] == 'Shot'].set_index('id')
    
    barcelona_shots = shots_match[shots_match['team'] == 'Barcelona']
    
    print('Match: ' + str(match_id) +': '+ team_required +' The Number of shots are: ' + str(len(barcelona_shots)))
```

```python
#Set number of matches
number_of_matches=ic+1

#Size of the pitch in yards (!!!)
pitchLengthX=120
pitchWidthY=80

pitch = Pitch()
# specifying figure size (width, height)
(fig, ax) = pitch.draw(figsize=(pitchLengthX, pitchWidthY))


#Plot the shots
#(fig,ax) = createPitch(pitchLengthX,pitchWidthY,'yards','gray')
for i,theshots in shots.iterrows():
    x=theshots['location'][0]
    y=pitchWidthY-theshots['location'][1]
    shotCircle=plt.Circle((x,y),1,color="blue")      
    shotCircle.set_alpha(.2)   
    ax.add_patch(shotCircle)

ax.set_title('Danger shots made by ' + team_required)
fig.set_size_inches(20, 14)

plt.show()

#Make x,y positions
x=[]
y=[]
for i,ashot in shots.iterrows():
    x.append(ashot['location'][0])
    y.append(pitchWidthY-ashot['location'][1])

#Make a histogram of shots
H_Shot=np.histogram2d(y, x,bins=5,range=[[0, pitchWidthY],[0, pitchLengthX]])


(fig, ax )= pitch.draw(figsize=(pitchLengthX, pitchWidthY))

pos=ax.imshow(H_Shot[0]/number_of_matches, extent=[0,120,0,80], aspect='auto',cmap=plt.cm.Reds)
fig.colorbar(pos, ax=ax)

plt.xlim((-1,121))
plt.ylim((83,-3))
plt.tight_layout()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
